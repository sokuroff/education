```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#     .xls
file_path = 'speed_data.xls'  #  'your_file.xls'     
df = pd.read_excel(file_path)

#     
time_interval = 0.005  #      
df['Time'] = df.index * time_interval

target_value = 250
upper_limit = target_value * 1.05
lower_limit = target_value * 0.95

#       x = 0.3
x_value = 0.299
y_value = np.interp(x_value, df['Time'], df['Mot.Spd.Fdb'])

#   
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 14

#  
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['Mot.Spd.Fdb'], linestyle='-', color = 'black', linewidth = '3')

plt.axhline(y=upper_limit, color='black', linestyle='--', label=f'+5%  {target_value}')
plt.axhline(y=lower_limit, color='black', linestyle='--', label=f'-5%  {target_value}')

#     x = 0.3,    
plt.plot([x_value, x_value], [0, y_value], color='black', linestyle='--', label='x = 0.3')

plt.xlabel(' ()')
plt.ylabel('   /')
plt.title(' ')
plt.grid(True)
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.show()
```
